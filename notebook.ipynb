Seminar 2 - 11.04.26
Analysis of the dataset with artists

```python
import csv
f = open("dataset.csv")
reader = csv.reader(f)
header = next(reader)

artist_i = header.index('artists')
artist_cnt = {}
popularity_i = header.index('popularity')
popularity_sum = {}
track_i = header.index('track_name')
unique_tracks = set()
top_tracks = []

for row in reader:
    artists = row[artist_i].split(';')
    popularity = int(row[popularity_i])
    track_name = row[track_i]
    first_artist = artists[0].strip()
    
    if track_name not in unique_tracks:
        unique_tracks.add(track_name)
        top_tracks.append((popularity, track_name, first_artist))
    
    for artist in artists:
        artist = artist.strip()
        if artist not in artist_cnt:
            artist_cnt[artist] = 0
            popularity_sum[artist] = 0
        artist_cnt[artist] += 1
        popularity_sum[artist] += popularity

f.close()

most_frequent = ''
max_cnt = 0
for artist in artist_cnt:
    if artist_cnt[artist] > max_cnt:
        max_cnt = artist_cnt[artist]
        most_frequent = artist
print("Most common artist:")
print(most_frequent, " with ", max_cnt, " tracks.")

most_popular = ''
least_popular = ''
max_avg = -1
min_avg = 1000000
for artist in artist_cnt:
    avg = popularity_sum[artist] / artist_cnt[artist]
    if avg > max_avg:
        max_avg = avg
        most_popular = artist
    if avg < min_avg:
        min_avg = avg
        least_popular = artist
print("\nMost popular artist:")
print(most_popular, " with ", max_avg, " percent of popularity.")
print("\nLeast popular artist:")
print(least_popular, " with ", min_avg, " percent of popularity.")

top_tracks.sort(reverse=True)
print("\nTop 10 tracks:")
for i in range(10):
    print(i + 1, top_tracks[i][1], "-", top_tracks[i][2], "-", top_tracks[i][0], "percent of popularity.")
    

```
